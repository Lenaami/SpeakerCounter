```python
#####
# Из файлика для загрузки выбрать только Array1-01
#####

import os

array = []

with open('amiBuild.txt', 'r') as f:
    text = f.readlines()

for txt in text:
    text_splt = txt.split('.')
    if text_splt[-2] == 'Array1-01':
        array.append(txt)

with open('amiBuild_Array1-01.txt', 'w') as f:
    f.writelines(array)

print('Количество файлов: ', len(array))
```

```python
#####
# Копирование файлов в другую папку
#####

import os

path_f = []
for d, dirs, files in os.walk('F:\\amicorpus_1'): # путь к файлам
    for f in files:
        path = os.path.join(d,f) # формирование адреса
        path_f.append(path) # добавление адреса в список
    
array = []

for path in path_f:
    array.append('COPY ' + path + ' F:\\amicorpus_2\n')

with open('copy_Array1-01.txt', 'w') as f:
    f.writelines(array)

print('Количество файлов для копирования: ', len(array))
```
